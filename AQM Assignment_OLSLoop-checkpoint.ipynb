```python
target_variables = ['BTC', 'BCH', 'LTC', 'ETH', 'EUR', 'CHF', 'GBP', 'RUB', 'YEN', 'SEK']
target_variables_log = ['BTC_log', 'BCH_log', 'LTC_log', 'ETH_log','EUR_log', 'CHF_log', 'GBP_log', 'RUB_log', 'YEN_log', 'SEK_log']

x = df[['Gold', 'Silver','Palladium','Platinum','OIL','MSCI', 'FTSE_100','NASDAQ','DAX', 'SMI']]
x = sm.add_constant(x)

#x = df[['Gold_log', 'Silver_log', 'Palladium_log', 'Platinum_log', 'OIL_log', 'MSCI_log', 'FTSE_100_log', 'NASDAQ_log', 'DAX_log', 'SMI_log']]
#x = sm.add_constant(x)

explanatory_variables = x.columns

results_of_interest = []
for i in explanatory_variables: results_of_interest.append('params.'+i)
for i in explanatory_variables: results_of_interest.append('bse.'+i)
for i in explanatory_variables: results_of_interest.append('pvalues.'+i)    
    
results_of_interest.extend(('rsquared','rsquared_adj','aic','autocorrelation', 'heteroscedasticity_lm', 'heteroscedasticity_lm_pvalue'))#, 'multicollinearity'))
```

```python
results_of_interest
```

```python
# list/array with all regression combinations
regression_combinations = []

explanatory_variables = ['Gold', 'Silver','Palladium','Platinum','OIL','MSCI', 'FTSE_100','NASDAQ','DAX', 'SMI']
for n in range(0, len(explanatory_variables)+1):
    regression_combinations.extend(list(itertools.combinations(explanatory_variables, n)))
```

```python
len(regression_combinations)
```

```python
#creation of all rows
rows = []
for asset in target_variables: 
    for i in range(1,len(regression_combinations)+1): rows.append(asset+str(i))
        
len(rows)
```

```python
# loop regression combination with rows

df_loop = pd.DataFrame(index=rows, columns=['y', 'x'])

for i in range(0,len(rows)):
    df_loop['y'][i] = target_variables[math.floor(i/len(regression_combinations))]
    df_loop['x'][i] = list(regression_combinations[int(math.remainder(i,len(regression_combinations)))])

df_loop
```
