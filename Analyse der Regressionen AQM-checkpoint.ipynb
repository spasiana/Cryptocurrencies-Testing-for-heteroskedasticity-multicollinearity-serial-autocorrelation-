# VIF calculation

```python
df_vif=pd.read_csv('AQSDataLog.csv', sep=',')
df_vif['Date'] = pd.to_datetime(df_vif['Date']) # change to datetime
df_vif.set_index('Date', inplace=True) #index 
df_vif = df_vif.rename(columns={'FTSE 100': 'FTSE_100', 'FTSE 100_log': 'FTSE_100_log'})
dflog_vif = df_vif.drop(['BTC','BCH','LTC', 'ETH','EUR', 'CHF', 'GBP', 'RUB', 'YEN', 'SEK', 'Gold', 'Silver','Palladium','Platinum','OIL',
                'MSCI', 'FTSE_100','NASDAQ','DAX', 'SMI'], axis=1)
dflog_vif.dropna(inplace=True)
```

```python
dflog_vif.head()
```

```python
print(dflog_vif.columns.values)
```

```python
# independent Variables
X = dflog_vif[['Gold_log', 'Silver_log', 'Palladium_log','Platinum_log', 'OIL_log',
          'MSCI_log', 'FTSE_100_log', 'NASDAQ_log', 'DAX_log', 'SMI_log']]
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
```
